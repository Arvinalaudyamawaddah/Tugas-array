```python
# perkalian matrix arrray

import numpy as np

A = np.array([[18, 12, 13]])
B = np.array([[20, 30, 40]])
C = np.array([[19, 18, 20]])

perkalian = A * B * C
print(perkalian)
```

```python
# pengurangan matrix array

import numpy as np

A = np.array([[18, 12, 13]])
B = np.array([[20, 30, 40]])
C = np.array([[19, 18, 20]])

pengurangan = A - B - C
print(pengurangan)
```

```python
# transpose matrix array

import numpy as np

A = np.array([[18, 12, 13], [20, 30, 40], [19, 18, 20]])

transpose_A = np.transpose(A)
print(transpose_A)
```

```python
# nilai invers

import numpy as np

A = np.array([[18, 12, 13], [20, 30, 40], [19, 18, 20]])

inverse_A = np.linalg.inv(A)
print("\nInvers matriks A:")
print(inverse_A)
```

```python
# nilai determinan 

import numpy as np

A = np.array([[18, 12, 13], [20, 30, 40], [19, 18, 20]])

determinan_A = np.linalg.det(A)
print("Determinan dari matriks A:", determinan_A)
```
